```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

df = pd.read_csv('data.csv')


group_by_diagnosis = df.loc[:, ['Diagnosis','Age','BloodPres','Chol']].groupby('Diagnosis')
avgs = group_by_diagnosis.mean()
x = avgs.index
y1 = avgs.Age
def plot(x, y, ax, title, y_label):
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.plot(x, y)
    ax.margins(x=0, y=0)

avgs.STPeakSlope.plot.bar()
#df.plot.bar()
```
